# Preprocessing

## Outlier Removal

```python
import numpy as np
print(f'Jumlah baris sebelum memfilter outlier: {len(df_train)}')

features = ['account_length', 
            'number_vmail_messages', 
            'total_day_minutes', 'total_day_calls', 'total_day_charge', 
            'total_eve_minutes', 'total_eve_calls', 'total_eve_charge',
            'total_night_minutes', 'total_night_calls', 'total_night_charge',
            'total_intl_minutes', 'total_intl_calls', 'total_intl_charge']

filtered_entries = np.array([True] * len(df_train))
for col in features:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - (IQR * 1.5)
    high_limit = Q3 + (IQR * 1.5)

    filtered_entries = ((df_train[col] >= low_limit) & (df_train[col] <= high_limit)) & filtered_entries
    
df_train = df_train[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(df_train)}')
```

## Normalization, Feature Selection, & Splitting

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
```

```python
X = df_train.drop(columns = 'churn')
y = df_train['churn']
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

features = ['account_length', 
            'international_plan', 'voice_mail_plan', 
            'number_vmail_messages', 
            'total_day_minutes', 'total_day_calls', 'total_day_charge', 
            'total_eve_minutes', 'total_eve_calls', 'total_eve_charge',
            'total_night_minutes', 'total_night_calls', 'total_night_charge',
            'total_intl_minutes', 'total_intl_calls', 'total_intl_charge',
            'number_customer_service_calls', 'area_code_area_code_408', 'area_code_area_code_415', 'area_code_area_code_510']

feature_select = SelectKBest(f_classif, k=6)
X_transform = scaler.fit_transform(X)
X_new = feature_select.fit_transform(X_transform, y)
features = [features[i] for i in feature_select.get_support(indices=True)]
features
```

```python
# Data Balancing

from imblearn.over_sampling import RandomOverSampler, SMOTE

#oversampler = RandomOverSampler(random_state=23)
oversampler = SMOTE()

X_ros, y_ros = oversampler.fit_resample(X_new, y)

print('Original dataset shape', y.shape)
print('Resample dataset shape', y_ros.shape)
```

```python
#pd.DataFrame({'variable': features, 'score_f_anova': feature_select.scores_})
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_ros, y_ros, test_size = 0.3, random_state = 42)
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.neighbors import NearestNeighbors


from sklearn.metrics import confusion_matrix, classification_report 
from sklearn.metrics import roc_auc_score, accuracy_score,f1_score, precision_score, recall_score
```

```python
model_1 = LogisticRegression()
model_2 = KNeighborsClassifier(n_neighbors = 8)
model_3 = DecisionTreeClassifier(max_depth = 5, criterion = 'entropy')
model_4 = RandomForestClassifier(n_estimators = 30, criterion = 'gini', max_depth = 7)

model_1.fit(X_train, y_train)
model_2.fit(X_train, y_train)
model_3.fit(X_train, y_train)
model_4.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_1 = model_1.predict(X_test)
y_pred_2 = model_2.predict(X_test)
y_pred_3 = model_3.predict(X_test)
y_pred_4 = model_4.predict(X_test)
```

```python
print('MODEL 1 (Logistic Regression) :')
print(classification_report(y_test, y_pred_1))
print('MODEL 2 (KNN) :')
print(classification_report(y_test, y_pred_2))
print('MODEL 3 (Decision Tree) :')
print(classification_report(y_test, y_pred_3))
print('MODEL 4 (Random Forest) :')
print(classification_report(y_test, y_pred_4))
```

```python
plt.rcParams['figure.figsize'] = [15, 15]

confusion_matrix_1 = pd.DataFrame((confusion_matrix(y_test, y_pred_1)), ('Not Churn', 'Churn'),('Not Churn', 'Churn'))
confusion_matrix_2 = pd.DataFrame((confusion_matrix(y_test, y_pred_2)), ('Not Churn', 'Churn'),('Not Churn', 'Churn'))
confusion_matrix_3 = pd.DataFrame((confusion_matrix(y_test, y_pred_3)), ('Not Churn', 'Churn'),('Not Churn', 'Churn'))
confusion_matrix_4 = pd.DataFrame((confusion_matrix(y_test, y_pred_4)), ('Not Churn', 'Churn'),('Not Churn', 'Churn'))

fig, axes = plt.subplots(2,2)
sns.heatmap(confusion_matrix_1, annot=True, ax = axes[0,0], annot_kws={'size': 14}, fmt='d', cmap="YlGnBu")
sns.heatmap(confusion_matrix_2, annot=True, ax = axes[0,1], annot_kws={'size': 14}, fmt='d', cmap="YlGnBu")
sns.heatmap(confusion_matrix_3, annot=True, ax = axes[1,0], annot_kws={'size': 14}, fmt='d', cmap="YlGnBu")
sns.heatmap(confusion_matrix_4, annot=True, ax = axes[1,1], annot_kws={'size': 14}, fmt='d', cmap="YlGnBu")

axes[0,0].set_title('Confusion Matrix for Logistic Regression Model')
axes[0,1].set_title('Confusion Matrix for KNN Model')
axes[1,0].set_title('Confusion Matrix for Decision Tree Model')
axes[1,1].set_title('Confusion Matrix for Random Forest Model')

for i in list(range(2)):
  for j in list(range(2)):
    axes[i,j].set_xlabel('Predicted Label')
    axes[i,j].set_ylabel('True Label')
```

```python
acc_comp = [
    accuracy_score(y_test, y_pred_1),
    accuracy_score(y_test, y_pred_2),
    accuracy_score(y_test, y_pred_3),
    accuracy_score(y_test, y_pred_4)
]

prec_comp = [
    precision_score(y_test, y_pred_1),
    precision_score(y_test, y_pred_2),
    precision_score(y_test, y_pred_3),
    precision_score(y_test, y_pred_4)
]

rec_comp = [
    recall_score(y_test, y_pred_1),
    recall_score(y_test, y_pred_2),
    recall_score(y_test, y_pred_3),
    recall_score(y_test, y_pred_4)
]

f1_comp = [
    f1_score(y_test, y_pred_1),
    f1_score(y_test, y_pred_2),
    f1_score(y_test, y_pred_3),
    f1_score(y_test, y_pred_4)
]

idx = ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest']

compare = pd.DataFrame({'Accuracy': acc_comp, 'Precision': prec_comp, 'Recall': prec_comp, 'F1': f1_comp}, index = idx)
compare.sort_values(by = ['Accuracy', 'Precision', 'Recall', 'F1'], ascending = False)
```
